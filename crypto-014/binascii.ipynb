```python
def str2bin(string):
    return bin(int(string.encode('hex'), 16))[2:]
```

```python
import string
def xor_str():
    for i in string.uppercase:
        print '{}\t'.format(i), str2bin(i)
        print "' '\t", '0' + str2bin(' ')
        print "XOR\t", bin(int(i.encode('hex'), 16) ^ int(' '.encode('hex'), 16))[2:]
        print '\n'
        
xor_str()
```
